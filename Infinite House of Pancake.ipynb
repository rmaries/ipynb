```python
def is_odd(num):
    return num & 0x1
def no_of_minutes(max_p):
    if max_p <= 2:
        return max_p
    elif is_odd(max_p):
        return 1+no_of_minutes((max_p+1)/2)
    else:
        return 1+no_of_minutes((max_p)/2)
nm = {}
for i in range(1,1001):
    nm[i] = no_of_minutes(i)
def tot_min(p_list):
    min_time = 0
    for j in p_list:
        if j > 2:
            min_time += nm[j]
    return min_time
#input_file = open("input.in", "r")
#n = int(input_file.readline())
#for line in range(n):
 #   inp_list = map(int,(input_file.readline()).split())
 #  print "Case #"+str(line+1)+": "+str(tot_min(inp_list))
```
