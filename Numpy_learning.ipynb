```python
%matplotlib inline
import pylab
pylab.plot(np.sin(a))
```

```python
x = np.linspace(0, 2*np.pi, 50)
pylab.plot(np.sin(x))
```

```python
pylab.plot(x, np.sin(x),
           x, np.sin(2*x))
```

```python
pylab.plot(x, np.sin(x), 'r-^')
```

```python
pylab.plot(x, np.sin(x), 'b-o',
     x, np.sin(2*x), 'r-^')
```

```python
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
pylab.scatter(x, y)
```

```python
#marker size/color set with data
x = np.random.rand(200)
y = np.random.rand(200)
size = np.random.rand(200)*30
color = np.random.rand(200)
pylab.scatter(x, y, size, color)
pylab.colorbar()
```
